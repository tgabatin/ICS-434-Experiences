```python
# Import Statements
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
# Creation of Random Seed
random.seed(42)
```

```python
# Testing the Random Sampling
random.seed(42)
random_nums = []
for i in range (10):
    random_nums.append(random.randint(0,1000))
print(random_nums)
```

```python
students_affiliations = []
categories = ["CS", "Engineering", "Math", "Physics"]
student_affiliations = pd.Series(np.random.choice(categories, size = 1000)).value_counts()
student_affiliations
```

```python
plt.figure(figsize=(12,7))
plt.title('Distribution of Students in the Program', color='green')
plt.bar(student_affiliations.index, student_affiliations.values,  color='#FF5733', alpha=0.55)
plt.xlabel('Disciplines')
plt.ylabel('Counts')
```
